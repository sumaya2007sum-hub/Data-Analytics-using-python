```python
def get_grade(mark):
  if mark >= 90:
    return 'S'
  elif mark >= 75:
    return 'A'
  else:
    return 'F'
students = {
    "Rahul": 95,
    "Priya": 82,
    "Arun" : 65,
    "Divya" : 91,
    }
passed = 0
failed = 0

for name, mark in students.items():
    grade = get_grade(mark)
    print(name, "-",mark, "-", grade)

    if grade == 'F':
       failed += 1
    else:
       passed += 1
print("passed:" , passed)
print("Failed:" , failed)
```

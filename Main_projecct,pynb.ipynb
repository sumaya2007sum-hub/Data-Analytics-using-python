```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Employee Attrition Dataset
data = {
    'EmployeeID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', 'IT', 'HR'],
    'Age': [25, 30, 28, 35, 26, 40, 32, 29],
    'Salary': [25000, 50000, 35000, 60000, 28000, 45000, 55000, 30000],
    'Attrition': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes']
}

# Create DataFrame
df = pd.DataFrame(data)

# -------------------------------
# Data Cleaning
# -------------------------------
print("Missing Values:")
print(df.isnull().sum())

# Remove duplicate rows if any
df = df.drop_duplicates()

print("\nCleaned Dataset:")
print(df)

# -------------------------------
# GroupBy Analysis
# -------------------------------
dept_attrition = df.groupby('Department')['Attrition'].value_counts()
print("\nDepartment-wise Attrition:")
print(dept_attrition)

# Count attrition employees
attrition_count = df['Attrition'].value_counts()
print("\nOverall Attrition Count:")
print(attrition_count)

# -------------------------------
# Visualization - Pie Chart
# -------------------------------
plt.figure(figsize=(5,5))
attrition_count.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Employee Attrition Distribution')
plt.ylabel('')
plt.show()

# -------------------------------
# Visualization - Bar Chart
# -------------------------------
dept_attrition_unstack = df.groupby('Department')['Attrition'].value_counts().unstack()

dept_attrition_unstack.plot(
    kind='bar',
    figsize=(6,4)
)

plt.title('Department-wise Employee Attrition')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()
```
